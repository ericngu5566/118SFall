```python
# --- Block 1: Offline Customer Support Chat (Simulated AI) ---
# No API key or internet needed. Uses basic logic to simulate an AI assistant.

from IPython.display import clear_output
import random

print("ðŸ’¬ Offline Customer Support Chat")
print("Type 'exit' anytime to end.\n")

# A few example "AI" responses
general_responses = [
    "I'm sorry to hear that. Could you please tell me more about the issue?",
    "I understand your concern. Letâ€™s see what we can do to fix this.",
    "Thanks for bringing this up. Can you confirm when the problem started?",
    "I can help you with that. Letâ€™s try a quick troubleshooting step."
]

troubleshooting_responses = [
    "Please try restarting your device and see if that helps.",
    "Make sure your internet connection is stable.",
    "Check if any recent updates were installed that could cause this.",
    "Try logging out and back in to refresh your settings."
]

closing_responses = [
    "Iâ€™m glad we could sort that out!",
    "Thanks for your patience â€” happy to help anytime.",
    "Your issue seems resolved. Have a great day!",
    "Itâ€™s been a pleasure assisting you today!"
]

# Chat loop
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("\nðŸ‘‹ Chat ended. Have a great day!")
        break

    # Simple keyword-based branching
    if any(word in user_input.lower() for word in ["problem", "issue", "error", "broken", "doesn't work"]):
        ai_message = random.choice(general_responses)
    elif any(word in user_input.lower() for word in ["wifi", "internet", "connection", "slow"]):
        ai_message = "Let's check your connection. Are other devices working fine?"
    elif any(word in user_input.lower() for word in ["restart", "fix", "tried", "reboot"]):
        ai_message = random.choice(troubleshooting_responses)
    elif any(word in user_input.lower() for word in ["thanks", "thank you", "resolved", "works now"]):
        ai_message = random.choice(closing_responses)
    else:
        ai_message = "Got it. Could you please clarify a bit more so I can assist better?"

    print(f"Support Agent: {ai_message}\n")
```
