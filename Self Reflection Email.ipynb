```python
# --- Block 3: Self-Reflection Prompt for Improving Output (Generative Email Example) ---

import random
from google.colab import files

# Step 1: Define possible scenarios
scenarios = [
    {
        "context": "emailing a teacher for a homework extension",
        "recipient": "Professor Johnson",
        "subject": "Request for Extension on Homework",
        "reason": "personal reasons",
        "proposed_deadline": "Friday",
    },
    {
        "context": "emailing a manager to request a day off",
        "recipient": "Mr. Thompson",
        "subject": "Request for Time Off",
        "reason": "a family commitment",
        "proposed_deadline": "next Monday",
    },
    {
        "context": "emailing a club leader about missing a meeting",
        "recipient": "Ms. Carter",
        "subject": "Apology for Missing Meeting",
        "reason": "a scheduling conflict",
        "proposed_deadline": "the next session",
    }
]

# Step 2: Randomly select a scenario
scenario = random.choice(scenarios)

# Step 3: Generate an initial (basic) email draft
initial_email = f"""
Subject: {scenario['subject']}

Hi {scenario['recipient']},

I couldn't make it or complete the task because of {scenario['reason']}.
Can I fix it or make it up by {scenario['proposed_deadline']}?

Thanks,
Alex
"""

# Step 4: Self-reflection (AI critiques its own writing)
reflection = f"""
Reflection:
- The tone is polite but too casual for {scenario['context']}.
- The message lacks structure and clear reasoning.
- Needs a more professional greeting and closing.
- Could express gratitude and accountability more clearly.
"""

# Step 5: Improved version based on reflection
improved_email = f"""
Subject: {scenario['subject']}

Dear {scenario['recipient']},

I hope you are doing well. I wanted to reach out regarding {scenario['context']}.
Unfortunately, I was unable to attend or complete the task due to {scenario['reason']}.
I take full responsibility and would like to make it up by {scenario['proposed_deadline']} if possible.

Thank you very much for your understanding and support.

Sincerely,
Alex Rivera
"""

# Step 6: Display outputs
print("=== INITIAL EMAIL ===")
print(initial_email)
print("\n=== SELF-REFLECTION ===")
print(reflection)
print("\n=== IMPROVED EMAIL ===")
print(improved_email)

# Step 7: Save and download improved email
filename = "improved_email_reflection.txt"
with open(filename, "w") as f:
    f.write(improved_email)

files.download(filename)
```
